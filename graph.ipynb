```python
%%output size=150
%%opts Curve [width=300 height=300 show_frame=True tools=['hover']]
%%opts Curve (line_width=2 )[logx=True ]
%%opts Overlay [width=300 height=300 legend_position='top_left' show_title=True]
#%%output filename="holoviews" fig="html"

screenclass = np.array( [ 2.0**0, 2.0**1, 2.0**2, 2.0**3, 2.0**4, 2.0**5, 2.0**6] )
dsize = np.array( [ 10**(0.5*(np.log10(screenclass[i]) + np.log10(screenclass[i+1]))) for i in range(len(screenclass)-1) ] )
    
f = r'https://raw.githubusercontent.com/computational-sediment-hyd/1DBedVariationMixed/master/all.json'
with urllib.request.urlopen(f) as url:
    d = json.loads(url.read().decode())
    
# d = json.load( open('all.json', 'r') )
nx = len(d[0]['data'])
nt = len(d)
dx = 100

def curves(t, x):
    da = np.array(d[t]['data'][x]['ratio'])
    da = np.add.accumulate( da )
    da = np.insert(da,0,0)
    a = hv.Curve((screenclass, da), extents = (1,0,100,1), label='output')
    
    da = np.array(d[0]['data'][0]['ratio'])
    da = np.add.accumulate( da )
    da = np.insert(da,0,0)
    b = hv.Curve((screenclass, da), extents = (1,0,100,1), label='first')

    return hv.Overlay([a,b]).relabel(str(d[t]['time'])+'sec '+str(x*dx)+'m     ')

dmap = hv.DynamicMap(curves, kdims=['t_index','x_index'])
dmap.redim.values(t_index=range(nt)).redim.values(x_index=range(nx))
```

```python
def fm_modulation(f_carrier=220, f_mod=220, mod_index=1, length=0.1, sampleRate=2000):
    sampleInc = 1.0/sampleRate
    x = np.arange(0,length, sampleInc)
    y = np.sin(2*np.pi*f_carrier*x + mod_index*np.sin(2*np.pi*f_mod*x))
    return hv.Curve((x, y), 'Time', 'Amplitude')
```

```python
%%opts Curve [width=600]
f_carrier = np.linspace(20, 60, 50)
f_mod = np.linspace(20, 100, 50)

curve_dict = {(fc, fm): fm_modulation(fc, fm) for fc in f_carrier for fm in f_mod}

# kdims = [hv.Dimension(('f_carrier', 'Carrier frequency'), default=40),
#          hv.Dimension(('f_mod', 'Modulation frequency'), default=60)]
kdims = [hv.Dimension(('f_carrier', 'Carrier frequency')),
         hv.Dimension(('f_mod', 'Modulation frequency'))]
holomap = hv.HoloMap(curve_dict, kdims=kdims)

# %%opts NdOverlay [width=600 legend_position='right']
# holomap.overlay('f_carrier')
holomap
```
